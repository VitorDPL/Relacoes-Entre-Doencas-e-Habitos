-🍎 Fruit_Consumption: Esse recurso também está inclinado para a direita. Muitos pacientes consomem frutas regularmente, mas um número significativo as consome com menos frequência.

-🥦 Green_Vegetables_Consumption: distribuição normal. A maioria dos pacientes consomem vegetais moderadamente.

-🍟 FriedPotato_Consumption: Este recurso está distorcido para a direita. Muitos pacientes consomem batatas fritas com menos frequência, enquanto alguns as consomem com mais frequência.

#### verificando possíveis correlações entre valores

```python
disease_conditions = ['Heart_Disease', 'Skin_Cancer', 'Other_Cancer', 'Diabetes', 'Arthritis']
selected_variables = ['General_Health', 'Exercise', 'Sex', 'Age_Category', 'Smoking_History']

for disease in disease_conditions:
    for variable in selected_variables:
        plt.figure(figsize=(10, 4))
        sns.countplot(data=data, x=variable, hue=disease)
        plt.title('Relação entre ' + variable + ' e ' + disease)
        plt.xticks(rotation=90)
        plt.show()
```

Isto sugere que o exercício está associado a níveis mais saudáveis ​​de IMC. 🏋️‍♀️

A relação entre a categoria de IMC e as condições da doença (doença_coração, câncer_de pele, outro_câncer, diabetes, artrite) mostra os seguintes padrões:

Para ❤️ Doenças Cardíacas,    Diabetes e 💪 Artrite, a prevalência é maior entre aqueles com IMC “sobrepeso” e “obeso”. Isto sugere que estas condições podem estar associadas a níveis mais elevados de IMC.

Para 🌞 Skin_Cancer e 🦀 Other_Cancer, a prevalência parece estar distribuída de forma mais uniforme pelas diferentes categorias de IMC. Isto poderia sugerir que estes ticâncer cancro podem não estar tão fortemente associados ao IMC como as outras condições.ões.

#### mapeamentos

```python
# foi necessário importar o dataframe novamente.
data = pd.read_csv('df.csv')

diabetes_mapping = {
    'No' : 0,
    'No, pre-diabetes or borderline diabetes' : 0,
    'Yes' : 1,
    'Yes, but female told only during pregnancy' : 1
}

data['Diabetes'] = data['Diabetes'].map(diabetes_mapping)

data = pd.get_dummies(data, columns= ['Sex'])

binary_columns = ['Heart_Disease', 'Skin_Cancer', 'Other_Cancer', 'Depression', 'Arthritis', 'Smoking_History','Exercise']
for column in binary_columns:
    data[column] = data[column].map({'Yes' : 1, 'No' : 0})

general_health_mapping = {
    'Poor' : 1,
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Excellent': 4
}
data['General_Health'] = data['General_Health'].map(general_health_mapping)

checkup_mapping = {
    'Never': 0,
    '5 or more years ago': 0.2,
    'Within the past 5 years': 1,
    'Within the past 2 years': 2,
    'Within the past year': 3
}
data['Checkup'] = data['Checkup'].map(checkup_mapping)

age_category_mapping = {
    '18-24': 0,
    '25-29': 1,
    '30-34': 2,
    '35-39': 3,
    '40-44': 4,
    '45-49': 5,
    '50-54': 6,
    '55-59': 7,
    '60-64': 8,
    '65-69': 9,
    '70-74': 10,
    '75-79': 11,
    '80+': 12
}
data['Age_Category'] = data['Age_Category'].map(age_category_mapping)    
    
data.head()
```

```python
# removendo linhas duplicadas.
data = data.drop_duplicates()
```

```python
corr = data.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(20, 30))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.title('Mapa de correlação')
plt.show()
```

#### Correlação de cada característica com as variáveis ​​da doença


```python
disease_variables = ['Heart_Disease', 'Skin_Cancer', 'Other_Cancer', 'Diabetes']

corr = data.corr()

# Cria uma figura (fig) com uma grade de subplots (axs).
# uma figura com 4 subplots(len(disease_variables)
fig, axs = plt.subplots(len(disease_variables), 1, figsize=(5, 25))

for ax, disease in zip(axs, disease_variables):
    # excluindo correlações que comparam a doença com a própria doença.
    target_corr = corr[disease].drop(disease)

    target_corr_sorted = target_corr.sort_values(ascending=False)

    sns.heatmap(target_corr_sorted.to_frame(), cmap="coolwarm", annot=True, fmt='.2f', cbar=False, ax=ax)
    ax.set_title('Correlation with ' + disease)

plt.tight_layout()
plt.show()
```

🔎 Nos mapas de calor, podemos observar o seguinte:

❤️ Heart_Disease: Esta condição apresenta uma forte correlação positiva com Age_Category e General_Health, e uma correlação negativa com Exercise e Sex_Female.

🌞 Skin_Cancer: Esta condição está fortemente correlacionada positivamente com Age_Category e Sex_Male, e negativamente correlacionada com Sex_Female.

🦀 Other_Cancer: Esta condição apresenta uma forte correlação positiva com Age_Category e General_Health, e uma correlação negativa com Sex_Female.

🩸  Diabetes: Esta condição apresenta uma forte correlação positiva com Age_Category, General_Health e BMI, e uma correlação negativa com Exercise.
